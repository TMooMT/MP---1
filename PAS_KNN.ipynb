```python
df= pd.read_csv('/content/drive/MyDrive/heart.csv')
df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for sex , d in df.groupby('sex'):
    ax.scatter(d['age'], d['chol'], label= sex)

plt.legend(loc='upper left')
plt.title('Age and cholestrol data')
plt.xlabel('Age')
plt.ylabel('Cholestrol')
plt.grid(True)
plt.show()
```

```python
import numpy as np

X_train = np.array(df[['age', 'chol']])
y_train = np.array(df['sex'])

print(f'X_train:\n{X_train}\n')
print(f'y_train: {y_train}')
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print(f'y_train:\n{y_train}')
```

```python
y_train = y_train.flatten()
print(f'y_train: {y_train}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(X_train, y_train)
```

```python
Age = 155
Chol = 70
X_new = np.array([Age, Chol]).reshape(1, -1)
X_new
```

```python
y_new = model.predict(X_new)
y_new
```

```python
lb.inverse_transform(y_new)
```

```python
X_test = np.array([[57, 126], [71, 149], [57, 131], [44, 141]])
y_test = np.array(['1', '0', '1', '0'])

print(f'X_test:\n{X_test}\n')
print(f'y_test:\n{y_test}')
```

```python
y_test = y_test.astype(int)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)

print(f'Accuracy: {acc}')
```

```python
from sklearn.metrics import precision_score
prec = precision_score(y_test, y_pred)

print(f'Precission: {prec}')
```

```python
from sklearn.metrics import recall_score

rec = recall_score(y_test, y_pred)

print(f'Recall: {rec}')
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)

print(f'F1-score: {f1}')
```

```python
from sklearn.metrics import classification_report

cls_report = classification_report(y_test, y_pred)

print(f'Classification Report:\n{cls_report}')
```

```python
from sklearn.metrics import matthews_corrcoef

mcc = matthews_corrcoef(y_test, y_pred)

print(f'MCC: {mcc}')
```
